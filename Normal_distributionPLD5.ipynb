```python
# The heights of adult women are normally distributed
# with a mean height of 175 cm and a standard deviation of 10 cm.
# a) Calculate and plot the probability density function (PDF) of this distribution,
# b) Compute the probability density for a woman who is 185 cm tall.

# Importing the necessary Libraries
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Normal distribution
mean = 175
std_dev = 10

# Generate x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the y values for the Normal distribution
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean)/std_dev)**2)

# Plot the Normal distribution
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Height Distribution of Adult Women', color='blue')
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean} cm')
plt.title('Normal Distribution (mean = 175 cm, std dev = 10 cm)')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Calculating the probability density at height of 185 cm
height = 185
pdf_value = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((height - mean)/std_dev)**2)
print(f"Probability density at height {height} cm: {pdf_value:.5f}")
```
